# pyear Tutorial
This notebook demonstrates feature extraction including frequency-domain metrics.

```
from pyear.pipeline import extract_features
from pyear.utils.epoch_aggregator import slice_raw_to_epochs
from unitest.fixtures.mock_ear_generation import _generate_signal_with_blinks
import matplotlib.pyplot as plt

signal, annotations = _generate_signal_with_blinks(100.0, 30.0, 1)
blinks = slice_raw_to_epochs(signal, annotations, 100.0, 30.0)
features = extract_features(blinks, 100.0, 30.0, 1, features=["frequency"])
print(features.head())
plt.plot(signal)
plt.title("Synthetic eyelid trace")
plt.show()
```
